8. Write a function `my_fun_plotter(f, x)` where `f` is a lambda function and `x` is an array. The function should plot `f` evaluated at `x`. Remember to label the $x$- and $y$-axes.

```python
import matplotlib.pyplot as plt
import numpy as np

def my_fun_plotter(f, x):
    plt.plot(x, f(x))
    plt.xlabel("$x$")
    plt.ylabel("$y$")

my_fun_plotter(lambda x: np.sqrt(x) + np.exp(np.sin(x)), np.linspace(0, 2*np.pi, 100))
plt.show()
```

9. Write a function with `my_poly_plotter(n, x)` that plots the polynomials $p_{k}(x) = x^{k}$ for $k = 1, \ldots, n$. Make sure your plot has axis labels and a title.

```python
def my_poly_plotter(n, x):
    plt.title(f"Polynomials up to degree {n}")
    for i in range(1, n+1):
        my_fun_plotter(lambda x: x**i, x)
    plt.show()

my_poly_plotter(7, np.linspace(-1, 1, 200))
```

10. Write a function `my_sierpinski(n)` that generates the points $p_{i}$ for $i = 1, \ldots, n$. THe function should make a plot of the points using blue dots.

```python
import random

def midpoint(P, Q):
    return ((P[0]+Q[0])/2, (P[1]+Q[1])/2)

def sierpinski_next_point(p):
    P = [(0, 0), (0.5, 1/(2**0.5)), (1, 0)]
    return midpoint(p, P[random.randint(0,2)])

def my_sierpinski(n):
    xpoints = [0]
    ypoints = [0]

    current_point = (xpoints[-1], ypoints[-1])
    for i in range(1, n):
        current_point = sierpinski_next_point(current_point)
        xpoints.append(current_point[0])
        ypoints.append(current_point[1])

    plt.scatter(xpoints, ypoints, s = 0.5, color = 'blue')
    plt.show()

my_sierpinski(10000)
```

11. Write a function `my_fern(n)` that generates the points $(x_{i},y_{i})$ for $i = 1, \ldots, n$ and plots them using blue dots. Also use `plt.axis("equal")` and `plt.axis("off")` to make the plot more attractive.

```python
import random

def fern_next_point(P):
    n = random.randint(1, 100)
    x = P[0]
    y = P[1]
    if n <= 85:
        return (0.85*x + 0.04*y, -0.04*x + 0.85*y + 1.6)
    elif n <= 92:
        return (-0.15*x + 0.28*y, 0.26*x + 0.24*y + 0.44)
    elif n <= 99:
        return (0.2*x - 0.26*y, 0.23*x + 0.22*y + 1.6)
    else:
        return (0, 0.16*y)

def my_fern(n):
    xpoints = [0]
    ypoints = [0]

    current_point = (xpoints[-1], ypoints[-1])
    for i in range(1, n):
        current_point = fern_next_point(current_point)
        xpoints.append(current_point[0])
        ypoints.append(current_point[1])

    plt.axis("equal")
    plt.axis("off")
    plt.title(f"My Fern with {n} Iterations")
    plt.scatter(xpoints, ypoints, s = 0.5, color = 'blue')
    plt.show()
    
my_fern(10000)
```
